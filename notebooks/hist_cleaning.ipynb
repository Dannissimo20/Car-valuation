```python
cars_config = cars_config.drop(
    columns=[
        'Рестайлинг', 
        'Ссылка на комплектацию', 
        'Год_low', 
        'Год_upp', 
        'Год_комплектация_low', 
        'Год_комплектация_upp', 
        'Список комплектаций'
    ]
)
```

```python
cars_config.head(5)
```

```python
cars['key'] = (
    cars['Метка'].astype(str) + '|' +
    cars['Объем двигателя'].astype(str) + '|' +
    cars['Мощность'].astype(str) + '|' +
    cars['Тип двигателя'].astype(str) + '|' +
    cars['Коробка передач'].astype(str) + '|' +
    cars['Привод'].astype(str) + '|' +
    cars['Руль'].astype(str) + '|' +
    cars['Тип кузова'].astype(str)
)

cars_config['key'] = (
    cars_config['Метка'].astype(str) + '|' +
    cars_config['Объем двигателя'].astype(str) + '|' +
    cars_config['Мощность'].astype(str) + '|' +
    cars_config['Тип двигателя'].astype(str) + '|' +
    cars_config['Коробка передач'].astype(str) + '|' +
    cars_config['Привод'].astype(str) + '|' +
    cars_config['Руль'].astype(str) + '|' +
    cars_config['Тип кузова'].astype(str)
)

lookup = (
    cars_config.drop_duplicates('key')
      .set_index('key')['Комплектация']
      .to_dict()
)


cars['Комплектация'] = cars['key'].map(lambda x: lookup.get(x))
cars = cars.drop(columns=['key', 'Метка'])
```

```python
split_df = cars['Название машины'].str.split(n=1, expand=True)
cars['Марка'] = split_df[0]
cars['Модель'] = split_df[1]
```

```python
cars[['Название машины', 'Марка', 'Модель', 'Комплектация', 'Объем двигателя', 'Мощность', 'Привод', 'Коробка передач', 'Тип двигателя', 'Тип кузова']].iloc[1100000:].head(20)
```

```python
drop_marks = ['Aston', 'Maserati', 'Bugatti', 'Ferrari', 'Lamborghini']
cars = cars[~cars['Марка'].isin(drop_marks)]
```

```python
cars.describe()
```

```python
cars.sort_values('Пробег', ascending=True)
```

```python
lower = cars['Цена'].quantile(0.15)
upper = cars['Цена'].quantile(0.85)
cars = cars[cars['Цена'] > lower]
cars = cars[cars['Цена'] < upper]
```

```python
cars.info()
```

```python
cars = cars.rename(columns={
    'Название машины': 'car_name',
    'Год': 'year',
    'Дата размещения объявления': 'date_posted',
    'Цена': 'price',
    'Объем двигателя': 'engine_volume',
    'Тип двигателя': 'engine_type',
    'Мощность': 'power',
    'Коробка передач': 'transmission',
    'Привод': 'drive',
    'Пробег': 'mileage',
    'Руль': 'steering_wheel',
    'Поколение': 'generation',
    'Рестайлинг': 'restyling',
    'Цвет': 'color',
    'Комплектация': 'trim',
    'Город': 'city',
    'Модель': 'model',
    'Марка': 'marka',
    'Тип кузова': 'body_type'
})
```

```python
cars['date_posted'] = pd.to_datetime(cars['date_posted'])
cars['year_month'] = cars['date_posted'].dt.to_period('M').dt.to_timestamp()
monthly_counts = cars.groupby('year_month').size()
monthly_counts_k = monthly_counts / 1000
monthly_counts_k = monthly_counts_k[monthly_counts_k >= 5]
monthly_counts_k
```

```python
from matplotlib import pyplot as plt


monthly_counts_k.plot()
plt.title('Monthly Listings Count (in thousands)')
plt.xlabel('Month')
plt.ylabel('Listings (K)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
cars = cars[cars['year_month'] > '2017-07-01']
```

```python
cars['generation'] = cars['generation'].astype('Int64')
cars['restyling'] = cars['restyling'].astype('Int64')
cars['year'] = cars['year'].astype('Int64')
```
