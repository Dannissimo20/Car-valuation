```python
import pandas as pd
```

```python
cars = pd.read_csv('../datasets/ml_raw_dataset.csv')
```

```python
cars.info()
```

```python
cars.describe()
```

Удаляем ненужные столбцы (дата размещения, ссылка на объявление, описание (сложно будет представить его для модели), дата спарсивания объявления)

```python
cars = cars.drop(columns=['date', 'link', 'description', 'parse_date', 'vehicleConfiguration', 'engineName'])
```

Проверка года выпуска автомобиля

```python
cars = cars.sort_values(by='year')
cars.head(10)
```

Оставляем объявления с автомобилями младше 1980 года

```python
cars = cars[cars['year'] > 1980]
cars.info()
```

Посмотрим редкость брендов на площадке

```python
cars_count = cars.groupby('brand')['year'].count().sort_values(ascending=False).reset_index(name='value')
cars_count
```

Оставим только те бренды которые представлены минимум в 300 экземплярах

```python
cars_count = cars_count[cars_count['value'] >= 300]
cars = cars[cars['brand'].isin(cars_count['brand'])]
cars.info()
```

Сконвертируем колонку engineDisplacement в float64

```python
cars = cars.dropna(subset=['engineDisplacement'])
cars['engineDisplacement'].unique()
```

```python
cars['engineDisplacement'] = cars['engineDisplacement'].str.replace(r'\s*LTR', '', regex=True).astype(float)
cars
```

Избавимся от строк которые имеют Nan в значимых столбцах

```python
cars.info()
```

```python
cars = cars.dropna(subset=['power', 'fuelType'])
cars.info()
```

Сконвертируем year в int

```python
cars['year'] = cars['year'].astype(int)
```

Создадим копию датасета с 710 000 строк

```python
cars_710k = cars
cars_710k['color'] = cars_710k['color'].fillna(cars_710k['color'].mode()[0])
cars_710k.to_csv('../datasets/ml_710k.csv', index=False)
```

Удалим все строки с nan пробегом и сохраним в отдельную копию датасета

```python
cars_410k = cars
cars_410k = cars_410k.dropna(subset=['mileage'])
cars_410k.to_csv('../datasets/ml_410k.csv', index=False)
```
